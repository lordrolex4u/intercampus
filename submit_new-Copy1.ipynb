```python
x = train_set.drop("Promoted_or_Not", axis=1)
y = train_set["Promoted_or_Not"]
```

```python
x = train_set.drop(['EmployeeNo', "Promoted_or_Not"], axis=1)
test  = test_set.drop('EmployeeNo', axis=1).copy()
x.shape, y.shape, test.shape
```

```python
x = pd.get_dummies(x, prefix_sep='_', drop_first=True)
test = pd.get_dummies(test, prefix_sep='_', drop_first=True)
```

```python
#splitting the dataset into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 225)
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
light_model = LGBMClassifier()
light_model.fit(x_train,y_train)

# make predictions for test set
y_pred = light_model.predict(x_test)
predictions = [round(value) for value in y_pred]

accuracy = accuracy_score(y_test, predictions) 
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
##hyperparameter tuning
```

```python
param_grid = {'num_leaves':[7,15,31,63],
             'max_depth':[3,4,5,6,-1]}
```

```python
from sklearn.model_selection import GridSearchCV
grid_sercher = GridSearchCV(estimator = light_model,param_grid =param_grid, cv=5 , verbose = 1,
                           n_jobs = 4)
```

```python
grid_sercher.fit(x_train,y_train)
```

```python
grid_sercher.best_params_,grid_sercher.best_score_
```

```python
accuracy_score(y_test,grid_sercher.predict(x_test))
```
